### Data Cleaning

```python
CompanyName = df['CarName'].apply(lambda df : df.split(' ')[0])
df.insert(3,"CompanyName",CompanyName)
df.drop(['CarName'],axis=1,inplace=True)
df.head()
```

```python
df.CompanyName.unique()
```

```python
def replace_name(a,b):
    df.CompanyName.replace(a,b,inplace=True)

replace_name('madfda','mazda')
replace_name('Nissan','nissan')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')
```

### Data Visulaisation

```python
df.columns
```

```python
df["price"].describe()
```

```python
categorical_features = [feature for feature in df.columns if df[feature].dtypes == 'O']
categorical_features
```

```python
### Let us find the value counts for categorical features

for feature in categorical_features:
    df[feature].value_counts().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

1. Gas fueltype is more preferred over diesel
2. Toyota is the most preferred.
3. Sedan is the most preferred carbody.
4. Ohc is the most preferred enginetype.
5. Four cylindered df are most preferred.

```python
### Let us find out the relationship between categorical features and dependent feature

for feature in categorical_features:
    data = df.copy()
    data.groupby(feature)['price'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("price")
    plt.title(feature)
    plt.show()
```

1. Diesel df have highest price
2. Jaguar company has the highest price
3. Obhcv has the highest price.

### Thus, we can say that price is inversely proportional to the number of sales.

```python
### Let us visualise numerical variables
numerical_features = [feature for feature in df.columns if df[feature].dtype!='O']
numerical_features
```

```python
for feature in numerical_features:
    if feature!='price':
        plt.scatter(df[feature],df['price'])
        plt.title(feature+ " vs price")
        plt.xlabel(feature)
        plt.ylabel("price")
        plt.show()
```

1. Carwidth, Carlength, Curbweight have positive correlation with price.
2. enginesize, boreratio, horsepower, wheelbase - seem to have a significant positive correlation with price.
3. citympg, highwaympg - seem to have a significant negative correlation with price.

```python
df['fueleconomy'] = (0.55 * df['citympg']) + (0.45 * df['highwaympg'])
```

```python
df['price'] = df['price'].astype('int')
temp = df.copy()
table = temp.groupby(['CompanyName'])['price'].mean()
temp = temp.merge(table.reset_index(), how='left',on='CompanyName')
bins = [0,10000,20000,40000]
df_bin=['Budget','Medium','Highend']
df['dfrange'] = pd.cut(temp['price_y'],bins,right=False,labels=df_bin)
df.head()
```

#### Finally, the significant features are:  -
1. Fueltype
2. Carbody
3. Enginetype
4. Number of cylinders
5. Carwidth
6. Carlength
7. Curbweight
8. Enginesize
9. Boreratio
10. Horsepower
11. Wheelbase
12. FuelEconomy

```python
df = df[["fueltype","carbody","enginetype","cylindernumber","carwidth","carlength","curbweight","enginesize","boreratio","horsepower","wheelbase","fueleconomy","price"]]
df.head()
```

```python
###
df["fueltype"] = df["fueltype"].map({"gas":0,"diesel":1})
df["carbody"] = df["carbody"].map({"convertible":0,"hatchback":1,"sedan":2,"wagon":3,"hardtop":4})
df["enginetype"] = df["enginetype"].map({"dohc":0,"ohcv":1,"ohc":2,"l":3,"rotor":4,"ohcf":5,"dohcv":6})
df["cylindernumber"] = df["cylindernumber"].map({"four":0,"six":1,"five":2,"three":3,"twelve":4,"two":5,"eight":6})
```

```python
df.head()
```

```python
df["fueleconomy"].min()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model_lr = LinearRegression()
```

```python
model_lr.fit(X_train,y_train)
```

```python
y_pred = model_lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor()
```

```python
model_rf.fit(X_train,y_train)
```

```python
y1_pred = model_rf.predict(X_test)
```

```python
r2_score(y_test,y1_pred)
```
